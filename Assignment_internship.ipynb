## Dimension Reduction

Dimensionality reduction refers to reducing the number of variables under consideration. Higher the number of variables harder it is to visualize the training data. Sometimes, most of feature or variables are correlated and hence are redundant. Common techniques used from dimensionality reduction are principle component analysis, etc
